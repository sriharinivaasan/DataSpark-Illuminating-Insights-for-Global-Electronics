```python
import pymysql
import pandas as pd

# MySQL connection
mydb = pymysql.connect(
    host="localhost",
    user="root",
    password="Malar@1976#",
    database="guviproj2"
)

mycursor = mydb.cursor()
mycursor.execute("drop table datadf")
mycursor.execute("CREATE TABLE  datadf(table1 VARCHAR(60),field1 VARCHAR(60),description1 VARCHAR(60))")

# Insert data from the DataFrame into the table
insert_query = "INSERT INTO datadf (table1,field1,description1) VALUES (%s, %s, %s)"
for _, row in dfdata.iterrows():
    mycursor.execute(insert_query, tuple(row))

# Commit the transaction and close the connection
mydb.commit()



print("dfdata successfully inserted into DFDATA.")
```

```python
mycursor.execute("CREATE TABLE  exdf(Date1 VARCHAR(10),Currency1 VARCHAR(5),Exchange1 FLOAT)")

# 2Insert data from the DataFrame into the table
insert_query = "INSERT INTO exdf (Date1,Currency1,Exchange1) VALUES (%s, %s, %s)"
for _, row in dfex.iterrows():
    mycursor.execute(insert_query, tuple(row))

# Commit the transaction and close the connection
mydb.commit()
```

```python
mycursor.execute("CREATE TABLE  prdf(ProductKey1 INT,ProductName1 VARCHAR(100),Brand1 VARCHAR(100),Color1 VARCHAR(100),UnitCostUSD1 VARCHAR(10),UnitPriceUSD1 VARCHAR(50),SubcategoryKey1 INT,Subcategory1 VARCHAR(50),CategoryKey1 INT,Category1 VARCHAR(50))")

# 2Insert data from the DataFrame into the table
insert_query = "INSERT INTO prdf (ProductKey1,ProductName1,Brand1,Color1,UnitCostUSD1,UnitPriceUSD1,SubcategoryKey1,Subcategory1,CategoryKey1,Category1) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for _, row in dfpr.iterrows():
    mycursor.execute(insert_query, tuple(row))

# Commit the transaction and close the connection
mydb.commit()
```

```python
mycursor.execute("CREATE TABLE  customersdf(CustomerKey1 INT,Gender1 VARCHAR(10),Name1 VARCHAR(100),City1 VARCHAR(100),StateCode1 VARCHAR(100),State1 VARCHAR(50),ZipCode1 VARCHAR(50),Country1 VARCHAR(50),Continent1 VARCHAR(50),Birthday1 VARCHAR(50))")
# 4Insert data from the DataFrame into the table
insert_query = "INSERT INTO customersdf (CustomerKey1,Gender1,Name1,City1,StateCode1,State1,ZipCode1,Country1,Continent1,Birthday1) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for _, row in dfcustomers.iterrows():
    mycursor.execute(insert_query, tuple(row))

# Commit the transaction and close the connection
mydb.commit()
print("customers df to sql done")
```

```python
mycursor.execute("CREATE TABLE  salesdf(OrderNumber1 INT,LineItem1 INT,OrderDate1 VARCHAR(20),DeliveryDate1 VARCHAR(20),CustomerKey1 INT,StoreKey1 INT,ProductKey1 INT,Quantity1 INT,CurrencyCode1 VARCHAR(20))")
# 5Insert data from the DataFrame into the table
insert_query = "INSERT INTO salesdf (OrderNumber1,LineItem1,OrderDate1,DeliveryDate1,CustomerKey1,StoreKey1,ProductKey1,Quantity1,CurrencyCode1) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for _, row in dfsales.iterrows():
    mycursor.execute(insert_query, tuple(row))

# Commit the transaction and close the connection
mydb.commit()
print("sales df to sql done")
```
